<h1>CREDIT CARD FRAUD DETECTION</h1>

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('fraudTrain.csv')
```

```python
df.head()
```

<h3>PREPROCESSING</h3>

```python
df.columns
```

```python
df.drop(columns=['Unnamed: 0','cc_num', 'first', 'last', 'gender', 'street', 'city', 'state', 'zip','dob', 'trans_num','trans_date_trans_time' ],inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df["merchant"] = encoder.fit_transform(df["merchant"])
df["category"] = encoder.fit_transform(df["category"])
df["job"] = encoder.fit_transform(df["job"])
```

```python
x = df.iloc[:-1, :-1]  # Exclude the last row and last column for x
y = df.iloc[:-1, -1]   # Exclude the last row, but only select the last column for y
```

```python
x
```

```python
y
```

<H3>DATA SPLITING AND MODEL FITTING</H3>

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
lr.fit(xtrain,ytrain)
```

```python
pred=lr.predict(xtest)
```

```python
from sklearn.metrics import accuracy_score

acc=accuracy_score(pred,ytest)
```

```python
print(acc)
```

```python
import tkinter as tk
from tkinter import messagebox
import pandas as pd  # Assuming you have pandas for data manipulation


# Function to handle button click and predict fraud
def check_fraud():
    # Get values from entry widgets
    merchant = int(entry_merchant.get())
    category = int(entry_category.get())
    amt = float(entry_amt.get())
    lat = float(entry_lat.get())
    long = float(entry_long.get())
    city_pop = int(entry_city_pop.get())
    job = int(entry_job.get())
    unix_time = int(entry_unix_time.get())
    merch_lat = float(entry_merch_lat.get())
    merch_long = float(entry_merch_long.get())

    # Create a DataFrame with the input data
    input_data = pd.DataFrame({
        'merchant': [merchant],
        'category': [category],
        'amt': [amt],
        'lat': [lat],
        'long': [long],
        'city_pop': [city_pop],
        'job': [job],
        'unix_time': [unix_time],
        'merch_lat': [merch_lat],
        'merch_long': [merch_long]
    })

    # Call the function to predict fraud (replace 'predict_fraud' with your actual prediction function)
    is_fraud =lr.predict(input_data)

    # Display the result
    if is_fraud == 1:
        result_text.set("This transaction is flagged as FRAUD!")
    else:
        result_text.set("This transaction is not fraudulent.")

# Create the main window
root = tk.Tk()
root.title("Credit Card Fraud Detection")

# Create entry widgets
labels = ['Merchant', 'Category', 'Amount', 'Latitude', 'Longitude', 'City Population', 'Job', 'Unix Time',
          'Merchant Latitude', 'Merchant Longitude']

for i, label_text in enumerate(labels):
    tk.Label(root, text=label_text).grid(row=i, column=0, sticky='e')

# Create entry widgets
entry_merchant = tk.Entry(root)
entry_category = tk.Entry(root)
entry_amt = tk.Entry(root)
entry_lat = tk.Entry(root)
entry_long = tk.Entry(root)
entry_city_pop = tk.Entry(root)
entry_job = tk.Entry(root)
entry_unix_time = tk.Entry(root)
entry_merch_lat = tk.Entry(root)
entry_merch_long = tk.Entry(root)

# Arrange entry widgets in a grid
entries = [entry_merchant, entry_category, entry_amt, entry_lat, entry_long, entry_city_pop, entry_job,
           entry_unix_time, entry_merch_lat, entry_merch_long]

for i, entry_widget in enumerate(entries):
    entry_widget.grid(row=i, column=1)

# Create a button to check fraud
check_button = tk.Button(root, text="Check Fraud", command=check_fraud)
check_button.grid(row=len(labels), column=0, columnspan=2)

# Create a label to display the result
result_text = tk.StringVar()
result_label = tk.Label(root, textvariable=result_text, font=('Helvetica', 12), fg='red')
result_label.grid(row=len(labels) + 1, column=0, columnspan=2)

# Start the Tkinter event loop
root.mainloop()
```
